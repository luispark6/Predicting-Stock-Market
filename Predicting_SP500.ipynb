```python
import yfinance as yf
```

```python
s500 = yf.Ticker("^GSPC")
```

```python
s500 = s500.history(period="max") #queries all data since the index was created
```

```python
s500
```

```python
del s500["Dividends"] #deleting dividends column because not useful information
```

```python
del s500["Stock Splits"] #deletng stock splits column because not useful information
```

```python
s500["Tomorrow_Close"] = s500["Close"].shift(-1) #this creates a new column for each row
#and shows the next days closing price. This will be used to see if stock prices increased
#or decreased in one day
```

```python
s500
```

```python
s500["Increase_Money"] = (s500["Tomorrow_Close"]>s500["Close"]).astype(int) #creating a new column that 
#indicates if the stock prices goes up(1) the next day or goes down(0)
```
